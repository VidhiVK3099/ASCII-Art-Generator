```python
from PIL import Image
import numpy as np

# Step 1: Define an expanded ASCII character set based on brightness levels
ASCII_CHARACTERS = "@%#*+=-:. " # Expanded character set

# Step 2: Function to resize the image while maintaining aspect ratio
def adjust_image_size(original_image, target_width):
    original_width, original_height = original_image.size
    aspect_ratio = original_height / original_width
    target_height = int(aspect_ratio * target_width * 0.55)  # Adjusting for ASCII aspect ratio
    return original_image.resize((target_width, target_height))

# Step 3: Function to convert the image to grayscale
def convert_image_to_grayscale(input_image):
    return input_image.convert("L")

# Step 4: Function to map grayscale pixel values to ASCII characters
def convert_pixels_to_ascii(input_image, ascii_characters=ASCII_CHARACTERS):
    pixel_array = np.array(input_image)  # Convert image to a numpy array
    ascii_art_lines = []


    # Loop through each pixel and map its brightness to an ASCII symbol
    for pixel_row in pixel_array:
        ascii_row = [ascii_characters[pixel_value // 32] for pixel_value in pixel_row]  # Divide by 32 to scale the pixel range (0-255) to the length of the ASCII character set
        ascii_art_lines.append("".join(ascii_row))

    return "\n".join(ascii_art_lines)

# Main function to convert an image file to ASCII art
def generate_ascii_art_from_image(file_name, output_width):
    try:
        # Open the image file using PIL
        image = Image.open(file_name)

        # Resize and convert the image to grayscale
        resized_image = adjust_image_size(image, output_width)
        grayscale_image = convert_image_to_grayscale(resized_image)

        # Convert pixels to ASCII characters
        ascii_art = convert_pixels_to_ascii(grayscale_image)

        return ascii_art

    except Exception as e:
        print(f"Error: {e}")
        return None

# Get user input for the image file name and desired output size
if __name__ == "__main__":
    input_file = input("Enter the image file name (with extension): ")
    desired_width = int(input("Enter the desired output width: "))

    # Generate the ASCII art
    ascii_art_output = generate_ascii_art_from_image(input_file, desired_width)

    # Output the ASCII art
    if ascii_art_output:
        print(ascii_art_output)

        # to save the ASCII art to a text file
        with open("ascii_art_output.txt", "w") as output_file:
            output_file.write(ascii_art_output)
```
